```python
#Label_encoding
from sklearn.preprocessing import LabelEncoder
a=LabelEncoder()
df['Genre']=a.fit_transform(df['Genre'])
```

```python
#Elbow_method
from sklearn.cluster import KMeans
wcss=[]
for i in range(1,11):
  kmeans=KMeans(n_clusters=i,init='k-means++',random_state=40)
  kmeans.fit(df)
  wcss.append(kmeans.inertia_)
plt.figure(figsize=(10,5))
plt.plot(range(1,11),wcss)
plt.title('the Elbow Method')
plt.xlabel("Number of clusters")
plt.ylabel('WcSS')
plt.show()
```

```python
#k_means_clustering
kmeans=KMeans(n_clusters=4,init='k-means++',random_state=40)
y_kmeans=kmeans.fit_predict(df)
```

```python
y_kmeans
```

```python
kmeans.inertia_
```

```python
df1 = df.values
```

```python
#visualizing_the_clusters
plt.figure(figsize=(15,7))
plt.scatter(df1[y_kmeans==0,3],df1[y_kmeans==0,4],color='yellow',label='cluster1',s=50)
plt.scatter(df1[y_kmeans==1,3],df1[y_kmeans==1,4],color='grey',label='cluster2',s=50)
plt.scatter(df1[y_kmeans==2,3],df1[y_kmeans==2,4],color='blue',label='cluster3',s=50)
plt.scatter(df1[y_kmeans==3,3],df1[y_kmeans==3,4],color='green',label='cluster4',s=50)
#plt.scatter(x[y_kmeans==4,0],x[y_kmeans==4,1],color='orange',label='cluster5',s=50)
plt.scatter(kmeans.cluster_centers_[:,3],kmeans.cluster_centers_[:,4],color='red',label='Centroids',s=150,marker=',')
plt.grid(False)
plt.title('Clusters of customers')
plt.xlabel('CustomerID')
plt.ylabel('Spending_Score')
plt.legend()
plt.show()
```

```python
#adding_new_column_named_cluster
df['Cluster']=pd.DataFrame(y_kmeans)
df
```
